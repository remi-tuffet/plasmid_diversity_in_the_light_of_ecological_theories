### Length multimodality

```python
data_plasmid_2=pd.read_csv('observed_data/Plasmid_size_mobility_PTU_coli.tsv',sep='\t')
```

```python
ptu_list=np.unique(data_plasmid_1['Predicted_PTU'])
ptu_mean_size=[]
conjugation=[]

for ptu in ptu_list:
    tmp=data_plasmid_2[data_plasmid_2['Predicted_PTU']==ptu]
    tmp2=data_plasmid_1[data_plasmid_1['Predicted_PTU']==ptu]
    ptu_mean_size=ptu_mean_size+[np.mean(tmp['size'])]
    conjugation=conjugation+[tmp2['main_mobility'].iloc[0]]
```

```python
ptu_dict=pd.DataFrame({'PTU' : ptu_list,
          'size' : ptu_mean_size,
          'size_log10' : np.log10(ptu_mean_size),
          'mobility' : conjugation
})
```

```python
## Figures 
transparence=0.5
fig, axs= plt.subplots(1,1,figsize=(15,10),sharex=True)

axs.hist(ptu_dict["size_log10"],bins=30,color="tab:blue",alpha=0.5,edgecolor="black",density=True)
sns.kdeplot(data=ptu_dict,x='size_log10',alpha=1,color='tab:blue',bw_method=0.25)
axs.set_ylabel('Density',size=30,weight="bold")
axs.set_xlabel('Length (bp)',size=30,weight="bold")
axs.set_xticks(ticks=np.log10([1000,5000,10000,50000,100000,500000])) 
axs.set_xticklabels(labels=['1e3','5e3','1e4','5e4','1e5','5e5'])
axs.tick_params(axis='both', labelsize=20)

axs.set_xlim(2.9,5.8)
axs.set_ylim(0,3.7)

plt.tight_layout(rect=[0, 0, 1, 1])
plt.show()
```

### Size multimodality

```python
data_bird_2=pd.read_csv('observed_data/bird_data.csv',sep=',')
```

```python
species_list=np.unique(data_bird_2['species.vertlife'])
bird_mean_size_cm=[]
family=[]

for bird in species_list:
    tmp=data_bird_2[data_bird_2['species.vertlife']==bird]
    bird_mean_size_cm=bird_mean_size_cm+[np.mean(tmp['totallength']/10)]
    family=family+[tmp['family'].iloc[0]]
```

```python
bird_dict=pd.DataFrame({'species' : species_list,
          'length_log10_cm' : np.log10(bird_mean_size_cm),
          'family' : family
})
```

```python
## Figures 
transparence=0.5
fig, axs= plt.subplots(1,1,figsize=(15,10),sharex=True)

axs.hist(bird_dict["length_log10_cm"],bins=30,color="tab:orange",alpha=0.5,edgecolor="black",density=True)
sns.kdeplot(data=bird_dict,x='length_log10_cm',alpha=1,color='tab:orange',bw_method=0.25)
axs.set_ylabel('Density',size=30,weight="bold")
axs.set_xlabel('Size (cm)',size=30,weight="bold")
axs.set_xticks(ticks=np.log10([10,20,40,60,80,100,120])) 
axs.set_xticklabels(labels=[10,20,40,60,80,100,120])
axs.tick_params(axis='both', labelsize=20)

axs.set_xlim(0.8,2.2)
axs.set_ylim(0,3.7)

plt.tight_layout(rect=[0, 0, 1, 1])
plt.show()
```
